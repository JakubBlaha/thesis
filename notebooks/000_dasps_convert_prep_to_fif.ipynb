```python
SECONDS_PER_EPOCH = None
SECONDS_PER_EPOCH = 5

OVERLAP = 0.5
```

```python
import mne
import os
import h5py
import numpy as np

DASPS_RAW_PATH = os.path.abspath("../data/dasps/raw_mat")
DASPS_CHNAMES = ['AF3', 'F7', 'F3', 'FC5', 'T7',
                 'P7', 'O1', 'O2', 'P8', 'T8', 'FC6', 'F4', 'F8', 'AF4']
DASPS_FS = 128

FIF_DATA_PATH = os.path.abspath("../data/fif")

def get_epochs_from_mat(fname):
    fname = os.path.join(DASPS_RAW_PATH, fname)

    with h5py.File(fname, 'r') as f:
        data = np.array(f['data'])

    info = mne.create_info(
        ch_names=DASPS_CHNAMES, ch_types="eeg", sfreq=DASPS_FS)
    info.set_montage('standard_1020')

    if SECONDS_PER_EPOCH:
        data = np.concatenate(data, axis=0)
        data = np.swapaxes(data, 0, 1)

        raw = mne.io.RawArray(data, info)
        epochs = mne.make_fixed_length_epochs(raw, duration=SECONDS_PER_EPOCH, overlap=OVERLAP * SECONDS_PER_EPOCH, preload=True)
    else:
        data = data.swapaxes(1, 2)

        epochs = mne.EpochsArray(data, info)

    epochs = epochs.filter(l_freq=4, h_freq=30)

    return epochs

def save_to_fif(fname, epochs):
    fname, _ = os.path.splitext(fname)
    fpath = os.path.join(FIF_DATA_PATH, fname + "-epo.fif")
    epochs.save(fpath, overwrite=True)

os.makedirs(FIF_DATA_PATH, exist_ok=True)

for fname in os.listdir(DASPS_RAW_PATH):
    epochs = get_epochs_from_mat(fname)
    fname = "dasps_prep_" + ("default_dur_" if SECONDS_PER_EPOCH is None else str(SECONDS_PER_EPOCH) + "s_dur_") + fname

    save_to_fif(fname, epochs)

    # break

epochs.plot(scalings="auto")
```
