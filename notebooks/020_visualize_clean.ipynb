```python
%matplotlib qt
import mne, os, glob, pandas as pd, re

def get_post_ar_path(pre_ar_path):
    _, fname = os.path.split(pre_ar_path)
    return os.path.join("../data/fif", "ar_" + fname) 

pre_ar = glob.glob("../data/fif/dasps_prep_*")
post_ar = [get_post_ar_path(fname) for fname in pre_ar]

zipped = list(zip(pre_ar, post_ar))

d = {"fnames": [], "pre": [], "post": [], "removed": [], "dur": []}

for pre_ar_path, post_ar_path in zipped:
    pre = mne.read_epochs(pre_ar_path)
    post = mne.read_epochs(post_ar_path)

    d["fnames"].append(pre_ar_path)
    d["pre"].append(len(pre))
    d["post"].append(len(post))
    d["removed"].append(len(pre) - len(post))

    duration = re.search(r"(\d+)s_dur_", pre_ar_path).group(1)
    duration = int(duration)
    d["dur"].append(duration)


df = pd.DataFrame(d)

df
```

```python
def print_stats_for_dur(seconds):
    df_filt = df[df["dur"] == seconds]

    total_pre = df_filt["pre"].sum()
    total_post = df_filt["post"].sum()
    total_removed = df_filt["removed"].sum()

    print(f"Total pre: {total_pre}")
    print(f"Total post: {total_post}")
    print(f"Total removed: {total_removed}")
    print(f"Removed: {total_removed / total_pre * 100:.2f}%")

print_stats_for_dur(5)
```

```python
def viz_first():
    viz = zipped[0]

    pre = mne.read_epochs(viz[0])
    post = mne.read_epochs(viz[1])

    pre.plot(scalings=100)
    post.plot(scalings=100)
```
