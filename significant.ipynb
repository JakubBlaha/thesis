```python
from main import *

trials, labels = get_trials_and_labels()
```

```python
import scipy.stats as stats

normal = []
anx = []

for trial, label in zip(trials, labels):
    if label == 0:
        normal.append(trial)
    else:
        anx.append(trial)

feat_names = normal[0].features.keys()

normal = np.asarray([np.asarray([*i.features.values()]) for i in normal])
anx = np.asarray([np.asarray([*i.features.values()]) for i in anx])

res = stats.f_oneway(normal, anx, axis=0)

pvalues = [
    (feat_name, pval) for feat_name, pval in zip(feat_names, res.pvalue)
]

pvalues = sorted(pvalues, key=lambda i: i[1])[::-1]
sig = [i for i in pvalues if i[1] < 0.05]

sig
```
